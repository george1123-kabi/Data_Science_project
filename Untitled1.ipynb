```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#---Data Generation---
np.random.seed(1)
customer_id=np.array([i for i in range(500)])
age=np.random.randint(18,70,size=500)
annual_income=np.random.randint(30000,150000,size=500)
total_spend=(annual_income*0.2)+np.random.randint(0,40000,size=500)


data={
    "CustomerID":customer_id,
    "Age":age,
    "AnnualIncome":annual_income,
    "TotalSpend":total_spend

}

df=pd.DataFrame(data)

#---Analysis---
#---Descriptive Statistics---
stats=df[['Age','AnnualIncome','TotalSpend']].describe()
print("Descriptive Statistics:\n",stats)

#---Creating Histogram---
plt.figure(figsize=(10,5))
sns.histplot(df['TotalSpend'],kde=True,color='teal')
plt.title("Distribution od Total Customer Spend")
plt.xlabel("Total Spend")
plt.ylabel("Frequency")
plt.show()

#---Creating Scatterplot---
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="AnnualIncome",y="TotalSpend")
sns.regplot(data=df,x="AnnualIncome",y="TotalSpend",scatter=False,color='red')
plt.title("Annual Income vs Total Spend")
plt.xlabel("Annual Income")
plt.ylabel("Total Spend")
plt.show()
```
